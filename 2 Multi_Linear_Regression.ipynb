```python
import pandas as pd
```

```python
dataset=pd.read_csv("50_Startups.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset)
```

```python
dataset
```

```python
dataset=dataset.astype(int)
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['R&D Spend', 'Administration', 'Marketing Spend','State_California', 'State_Florida', 'State_New York']]
```

```python
dependent=dataset[['Profit']]
```

```python
from sklearn.model_selection import train_test_split
train_test_split(independent, dependent, test_size=0.30, random_state=0)
```

```python
#Assign input train, input test, output train and output test variables.
x_train,x_test,y_train,y_test=train_test_split(independent, dependent, test_size=0.3, random_state=0)
```

```python
#Creating Model
#Importing LinearRegression Model from sklearn Library
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()

#Model creation
regressor.fit(x_train,y_train)
```

```python
#Creating Slope
weight=regressor.coef_
weight
```

```python
#Creating Origin
bais=regressor.intercept_
bais
```

```python
#Creating predicted output
y_pred=regressor.predict(x_test)
```

```python
#Creating Evalution Metrics
#importing r2_score from sklearn
from sklearn.metrics import r2_score
#Creating r_score value using given out and predicted output
r_score=r2_score(y_test,y_pred)
```

```python
r_score
```

```python
#Save Model using Pickle function
#Importing pickle will save all functions in pickle into memory
import pickle
```

```python
filename="Finalized_Model_Mul_Linear.sav"
```

```python
#Writing binary file into Python
pickle.dump(regressor,open(filename,'wb'))
```
